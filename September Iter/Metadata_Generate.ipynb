```python
import re
# Function to extract date
def extract_date(doc_name):
    # Match both YYYY.MM.DD and QX.YYYY formats
    match = re.search(r'(\d{4}\.\d{2}\.\d{2})|(Q\d\.\d{4})', doc_name)
    if match:
        return match.group(0)
    return None

# Apply the function to extract dates
main_df['Date'] = main_df['Document Name'].apply(extract_date)
```

```python
from datetime import datetime

def get_quarter_end_date(date):
    if date is None:
        return None

    if isinstance(date, str) and 'Q' in date:
        year = int(date.split('.')[1])
        quarter = int(date.split('.')[0][1])
    else:
        date_obj = datetime.strptime(date, '%Y.%m.%d')
        year = date_obj.year
        month = date_obj.month
        if month <= 3:
            quarter = 1
        elif month <= 6:
            quarter = 2
        elif month <= 9:
            quarter = 3
        else:
            quarter = 4

    # Return the quarter end date
    if quarter == 1:
        return f'{year}.03.31'
    elif quarter == 2:
        return f'{year}.06.30'
    elif quarter == 3:
        return f'{year}.09.30'
    elif quarter == 4:
        return f'{year}.12.31'

# Apply the function to the "date" column
main_df['quarter_end_date'] = main_df['Date'].apply(get_quarter_end_date)
import pandas as pd

main_df
```
